In this project we are going to build a decision tree classifier on Iris dataset.
The goal is if a new data entry is given ,the model should predict the target class accurately. 

**Importing necessary libraries to do our current project**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets as datasets
%matplotlib inline
```

**Loading the Iris dataset from datasets library**

```python
iris_data = datasets.load_iris()
```

**Converting the data into Pandas dataframe**

```python
df = pd.DataFrame(iris_data.data,columns=iris_data.feature_names)
```

Let us preview the shape and content of the iris dataset 

```python
print(df.shape)
df.head(10)
```

**Preview of target classes**

```python
y=iris_data.target
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train , y_test =train_test_split(df,y,test_size=0.25,random_state=0)
```

Now we have explored the data .Let us now create the Decsion Tree Classifier Model

**Declaring the model**

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
print("DECISION TREE CLASSIFIER model built sucessfuly")
```

**Let us visualize the built model graphically**

```python
#Instaling the required libraries 
!pip install pydotplus
!apt-get install graphviz -y
```

**Importing necessary libraries for graph viz**

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

Let us visualize the DECISION TREE GRAPHICALLY

```python
dot_data=StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris_data.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

Let us check how our model is performing on the new data

```python
y_predicted = dtree.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_predicted,y_test)
print("Accuracy {}".format(accuracy))
```

Our model is performing with good accuracy on the test data.
Any model with more than 90% accuracy is considered as a good Model.
Hence our model is fine to be deployed .
